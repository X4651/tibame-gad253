````python
# 三個步驟
import time
import urllib.request as req
import urllib.error as err
import bs4 as bs
import pandas as pd

def get_page(url):
    # url = "https://tabelog.com/tw/tokyo/rstLst/sweets/1/?SrtT=rt"
    while True:
        try:
            resp = req.urlopen(url)
            break
        except err.HTTPError:
            print("retry...")
            time.sleep(1)

    content = resp.read()
    html = bs.BeautifulSoup(content)

    # Step1. 準備一個空的list
    table = []

    rs = html.find_all("div", {"class":"list-rst__body"})
    for r in rs:
        en = r.find("a", {"class":"list-rst__rst-name-target"})
        rating = r.find("span", {"class":"c-rating__val"})
        area_genre = r.find("div", {"class":"list-rst__area-genre"})
        prices = r.find_all("span", {"class":"c-rating-v3__val"})
        dinner_price = prices[0]
        lunch_price = prices[1]
        holiday = r.find("span", {"class":"list-rst__holiday-text"})

        en_href = en["href"]
        en_text = en.text.strip()
        rating_text = rating.text.strip()
        area_genre_text = area_genre.text.strip()
        dinner_price_text = dinner_price.text.strip()
        lunch_price_text = lunch_price.text.strip()
        holiday_text = holiday.text.strip()

        # print(rating_text)
        # print(en_text)
        # print(en_href)
        # print(area_genre_text)
        # print(dinner_price_text)
        # print(lunch_price_text)
        # print(holiday_text)
        # print("-" * 30)
        # 評價 地點 晚餐價格 午餐價格 例假日

        # Step2. 把你的每一筆資料變成一個字典
        row = {
            "rating":rating_text,
            "en_name":en_text,
            "url":en_href,
            "area_genre":area_genre_text,
            "dinner_price":dinner_price_text,
            "lunch_price":lunch_price_text,
            "rest_day":holiday_text,
        }
        table.append(row)
    return table

total = []
for i in range(10):
    page = i + 1
    url = "https://tabelog.com/tw/okinawa/rstLst/{}/?SrtT=rt".format(page)
    print(url)
    partial = get_page(url)
    total = total + partial

# Step3. [] -> pd.DataFrame
df = pd.DataFrame(total)
df.to_csv("tabelog.csv", encoding="utf-8")
df
````

````python
def flow(area_genre):
    spl = area_genre.split(" / ")
    if len(spl) == 2:
        data = {"area":spl[0], "genre":spl[1]}
    else:
        data = {"area":"-", "genre":spl[0]}
    return pd.Series(data)

cols = ["area", "genre"]
df[cols] = df["area_genre"].apply(flow)
df
````

````python
# 對日期過濾轉換成表格 星期一到日變成七行 用01填在表格內
cols = [
    "Mon(open)",
    "Tues(open)",
    "Wed(open)",
    "Thu(open)",
    "Fri(open)",
    "Sat(open)",
    "Sun(open)"
]
def flow(s):
    # 先把全部的都先設成1，就不用做兩次
    dic = {c:1 for c in cols}
    trans = [
        "星期一",
        "星期二",
        "星期三",
        "星期四",
        "星期五",
        "星期六",
        "星期日",
    ]
    for i in range(len(trans)):
        refer = trans[i]
        if refer in s:
            dic[cols[i]] = 0
    return pd.Series(dic)

df[cols] = df["rest_day"].apply(flow)
df
````
