```python
# dumps 轉換成字串 dump會存在一個檔案
# read不能重複使用
# load loads
# dump dumps
import json
test = [
    {"name":"Elwing", "height":175},
    {"name":"Bob", "height":180}
]
json.dumps(test)
# !
with open("test.json", "w", encoding="utf-8") as f:
    json.dump(test, f)

with open("test.json", "r", encoding="utf-8") as f:
    test2 = json.load(f)
    print(test2)

# !
with open("test.json", "r", encoding="utf-8") as f:
    content = f.read()
    test3 = json.loads(content)
    print(test3)
```
