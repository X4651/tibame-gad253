```
HTML格式
- 由很多長方形(區塊)構成

<a href="xxx"> -> start
    店名
</a> -> end

1. 名字: 決定區塊用途
a: 超連結  img: 圖片  video: 影片

2. 屬性: 決定對應名字區塊細節(名="直")
a href="網址"
img/video src="網址"

3. 顯示文字


前端
1. 把回應呈現給你看
a. HTML: 內容(o)
b. CSS: 幫你把HTML做排版(o)
c. JavaScript: 幫你處理網頁的動態(x)

後端(ngrok+FastAPI):
1. 接收網址
2. 處理網址(查詢資料庫....)
3. 送回回應


CSS語法: 定位區塊 加上排版
1. 針對HTML名字做定位
a{
    background-color:pink
}

2. 特別HTML排版屬性
class="分類1 分類2 分類32"
.分類1{

}

e.g.
.js-thumbnail-img{
    border-radius:50%
}

3. 排版屬性
id="不重複的職"
```

````python
class Student:

    def __init__(self, h, w):
        self.height = h
        self.weight = w

    def bmi_sp(self):
        return self.weight / self.height

    def __add__(self, other):
        newh = (self.height + other.height) / 2.0
        neww = (self.weight + other.weight) / 2.0
        return Student(newh, neww)


s1 = Student(180, 77)
s1.bmi_sp()
s2 = Student(190, 80)

# s1.__add__(s2)
s3 = s1 + s2
print(s3.weight)
````

````python
import urllib.request as req
import bs4 as bs

url = "https://tabelog.com/tw/tokyo/rstLst/sweets/1/?SrtT=rt"
resp = req.urlopen(url)
content = resp.read()
html = bs.BeautifulSoup(content)

rs = html.find_all("div", {"class":"list-rst__body"})
for r in rs:
    en = r.find("a", {"class":"list-rst__rst-name-target"})
````
````python
    rating = r.find("span", {"class":"c-rating__val"})
    area_genre = r.find("div", {"class":"list-rst__area-genre"})
    prices = r.find_all("span", {"class":"c-rating-v3__val"})
    dinner_price = prices[0]
    lunch_price = prices[1]
    holiday = r.find("span", {"class":"list-rst__holiday-text"})

    en_href = en["href"]
    en_text = en.text.strip()
    rating_text = rating.text.strip()
    area_genre_text = area_genre.text.strip()
    dinner_price_text = dinner_price.text.strip()
    lunch_price_text = lunch_price.text.strip()
    holiday_text = holiday.text.strip()

    print(rating_text)
    print(en_text)
    print(en_href)
    print(area_genre_text)
    print(dinner_price_text)
    print(lunch_price_text)
    print(holiday_text)
    print("-" * 30)
    # 評價 地點 晚餐價格 午餐價格 例假日
````
