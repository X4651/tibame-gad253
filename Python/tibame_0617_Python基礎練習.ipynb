````python
# 優化後(簡化算式、方便修改數量跟符號)
left,right = " ","*"
layer = 10
i = 0
while i < layer:
    # 第一列九個空格一個符號
    lcount = layer -1 - i
    rcuont = i + 1
    print(left * lcount + right * rcuont)
    i = i + 1
````

````python
left,mid,right = " ","*"," "
layer = 10
i = 0
while i < layer:
    lcount = i
    # 19 17 15 13 11 9 7 5 3 1 間隔為2
    mcount = (layer * 2 - 1) - (2 * i)
    rcuont = i
    print(left * lcount + mid * mcount + right * rcuont)
    i = i + 1
````

````python
# 根據指令移動
x,y = 0,0
commands = "ULLRDL"
i = 0
while i < len(commands):
    c = commands[i]
    if c == "U":
        y = y + 1
    elif c == "D":
        y = y - 1
    elif c == "L":
        x = x - 1
    # 最後不能用else 不然輸入空格之類的也會執行 最後的else可省略
    elif c == "R":
        x = x + 1
    i = i + 1
print(x,y)
````

只要想不出while後面要寫甚麼，或者條件很複雜的，一律使用while True加上break的組合  
有三個優點:  
1.自由，要在哪做判斷，要在哪做break都可以，不用把條件都塞在while  
2.在break前要做任何收尾工作，不管是要列印或關掉檔案  
3.結束迴圈比想辦法讓他開始容易

````python
# 印出十位小數，增加整除時的提早停止機制
# 2 / 7 -> 0.2857142857
# 3 / 5 -> 0.6
a = 3
b = 5
ans = "0."
i = 0
while i < 10:
    # 2 -> 20
    a10 = a * 10
    # 20 // 7
    ans = ans + str(a10 // b)
    # print(ans)
    # 20 % 7
    a = a10 % b
    if a == 0:
        break
    i = i + 1
print(ans)
````

````python
# 終極密碼 寫法3 在最前面就保證guess是對的(模組化)
import random
low,high = 1,100
ans = random.randint(low,high)

while True:
    while True:
        print(f"請輸入{low}到{high}之間的數字:")
        guess = int(input())
        # 偷懶寫法 簡化guess > low and guess < high
        if low < guess < high:
            break
        else:
            print("輸入錯誤")
    if guess < ans:
        print("猜小了")
        low = guess
    elif guess > ans:
        print("猜大了")
        high = guess
    elif guess == ans:
        print("猜中了")
        break
````

````python
# list清單
# 正向key [0]  [1]  [2]  [3]  [4]
# 反向key[-5] [-4] [-3] [-2] [-1]
scores = [10 , 20 , 30 , 40 , 50]

print(sum(scores),max(scores),min(scores))
print(len(scores))
# 確認是否有在清單內
print(60 in scores)
# 查詢操作:要查的東西[key]
print(scores[0])
# 查最後一個:len(scores)會回傳5，但key只有0~4，所以要-1
# print(scores[len(scores)-1])
# 用反向key查詢最後一個，上面寫的並不pythonic(漂亮)
print(scores[-1])

# scores = [10 , 20 , 30 , 40 , 50]

# 查詢一段:要查的東西[開始key:結束key(不包含)] 含前不含後
# e.g.list[2:5] -> list[2] list[3] list[4]
print(scores[1:3])
# 如果要拿到最後一個的前一個 -1是指位置
print(scores[1:-1])

# 變種
# 拿012
print(scores[:3])
# 拿1234
print(scores[1:])
# 跳號:要查的東西[開始key:結束key(不包含):跳號]
# e.g.list[2:11:3] -> list[2] list[5] list[8]
print(scores[0:4:2])
# 相當於取奇數個
print(scores[::2])
# 取偶數個
print(scores[1::2])
# 反向跳號從右跳左(位置)
print(scores[::-2])
print(scores[4:0:-2])
````

````python
# 找出幾個及格
scores = [80,60,90,20,100]
i = 0
count = 0
while i < len(scores):
    # 如果key不是01234，這種寫法就會出事，要改用for in(走過)語法
    s = scores[i]
    if s >= 60:
        count = count + 1
    i = i + 1
print(count)
````

````python
# for in語法:for 名稱 in 群集(str/list...):
# 會保證每個東西都會拿給你一次，有順序的就會照順序，沒順序的就隨意
# 一定要先準備一個名稱在for跟in的中間

scores = [80,60,90,20,100]
count = 0
# s=80 s=60 s=90 s=20 s=100
for s in scores:
    if s >= 60:
        # 不用i，因為沒有在數次數
        count = count + 1
````

````python
# 一加到十，用while
i = 0
total = 0
while i < 10:
    total = total + (i + 1)
    i = i + 1
print(total)

# range跟list很像(含前不含後)
# 用法一:range(5)產生[0,1,2,3,4]
# 用法二:range(3,5)產生[3,4]
# 用法三:range(2,11,3)產生[2,5,8]

# 一加到十，用for in，數單純幾次的時候用
total = 0
# 產生一個0~9的list，每次都叫他i
for i in range(10):
    total = total + (i + 1)
print(total)
````
