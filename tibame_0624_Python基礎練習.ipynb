````python
# 這個函式庫用於處理網址相關的操作，例如下載檔案。
import urllib.request as req

# 圖片函式庫: opencv、Pillow(colab自帶，不用pip)
# 這個函式庫是 Python Imaging Library (PIL) 的一部分，
# 用於圖像處理，例如開啟和操作圖片檔案。
import PIL

# 這個函式庫用於進行高效能的數值運算，特別是處理陣列和矩陣。
import numpy as np

# 這個函式庫用於繪製圖形，包括顯示圖像。
import matplotlib.pyplot as plt

# 從指定的網址下載一張圖片
url = "https://github.com/Elwing-Chou/tibame_0618/raw/refs/heads/main/chiikawa.webp"
# 改檔名
req.urlretrieve(url, "chiikawa")

# 使用 PIL.Image.open() 函式開啟剛下載的圖片檔案
img = PIL.Image.open("chiikawa")

# 將 PIL 圖片物件 img 轉換成一個 NumPy 陣列，儲存在變數 img_np 中。
img_np = np.array(img)

# print(img_np)
# 高 寬 RGB
# print(img_np.shape)

# 如果確定裡面有三個值:
# r, g, b = img_np[0][0]
# 快速轉灰階
# img.convert("L")

# 練習轉灰階 灰階可能是0~255的整數或0~1的小數 看分子 綠色占比最重
# L = (R * 299 / 1000) + (G * 587 / 1000) + (B * 144 / 1000)
# .0確保是小數
img_np_float = img_np / 255.0
# 走過這個高/長
# 每一個位置本來是一個List [255, 128, 100]
# L = R * 299/1000 + G * 587/1000 + B * 114/1000
# 變成一個數設定回去(灰階強度)

# 這裡將高度儲存在變數 h 中，
# 寬度儲存在變數 w 中，顏色通道數儲存在變數 color 中。
h, w, color = img_np_float.shape
img_np_gray = np.zeros([h, w])

# 對於灰階: 0.0是純黑  1.0是純白
for i in range(h):
    for j in range(w):
        r, g, b = img_np_float[i][j]
        l = 0.299 * r + 0.587 * g + 0.114 * b
        # 二分法 不是黑就是白
        # if l > 0.5:
        #     l = 1.0
        # else:
        #     l = 0.0
        img_np_gray[i][j] = l
# 使用 matplotlib 函式庫顯示處理後的灰階圖片
# cmap 代表顏色的強度
plt.imshow(img_np_gray, cmap="gray")
# _r反轉光譜
# plt.imshow(img_np_gray, cmap="gray_r")
````

````python
# ?a?b
ans = "9229"
test = "9002"

count_a = 0
count_b = 0
# remain代表剩下的
ans_remain = {}
test_remain = {}

for i in range(len(ans)):
    c1, c2 = ans[i], test[i]
    if c1 == c2:
        count_a = count_a + 1
    else:
        if c1 in ans_remain:
            ans_remain[c1] = ans_remain[c1] + 1
        else:
            ans_remain[c1] = 1
        if c2 in test_remain:
            test_remain[c2] = test_remain[c2] + 1
        else:
            test_remain[c2] = 1
# print(ans_remain, test_remain)

# 走過test_remain裡面剩下的key
# 去看ans裡面還有幾個，來決定有幾個b
for key in test_remain:
    if key in ans_remain:
        # 0200 9992: 1b
        # 0200 9922: 1b
        # 計算 B 的數量，取 ans_remain 和 test_remain 中該字元數量的最小值
        count_b = count_b + min(ans_remain[key], test_remain[key])
print(f"{count_a}A{count_b}B")
````

````python
# ?a?b 完整版 含隨機答案、輸入數值、防呆
import random
DIGIT = 4

ans = ""
for i in range(DIGIT):
    n = random.randint(0, 9)
    ans = ans + str(n)


while True:
    while True:
        test = input("請猜:")
        if len(test) == DIGIT:
            break
        else:
            print("數量不相符合")

    count_a = 0
    ans_remain = {}
    test_remain = {}
    for i in range(len(ans)):
        c1, c2 = ans[i], test[i]
        if c1 == c2:
            count_a = count_a + 1
        else:
            if c1 in ans_remain:
                ans_remain[c1] = ans_remain[c1] + 1
            else:
                ans_remain[c1] = 1
            if c2 in test_remain:
                test_remain[c2] = test_remain[c2] + 1
            else:
                test_remain[c2] = 1
    # print(ans_remain, test_remain)

    # 走過test_remain裡面剩下的key
    # 去看ans裡面還有幾個, 來決定有幾個b
    # 0220 9992: 1B
    # 0200 9922: 1b
    count_b = 0
    for key in test_remain:
        if key in ans_remain:
            count_b = count_b + min(test_remain[key], ans_remain[key])

    print(count_a, "A", count_b, "B")

    if count_a == DIGIT:
        print("猜對了")
        break
````
